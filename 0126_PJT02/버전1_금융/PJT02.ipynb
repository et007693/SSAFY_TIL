```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./NFLX.csv')
```

```python
# A. 데이터 전처리 - 데이터 읽어오기
data = data[['Date','Open','High','Low','Close']]
data
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data = data[data['Date'].dt.year >= 2021]
```

```python
# B. 데이터 전처리 - 2021년 이후의 종가 데이터 출력
plt.plot('Date', 'Close', data=data)
plt.title('NFLS close Price')
plt.xlabel('Date')
plt.ylabel('Close price')
plt.show()
```

```python
# C. 데이터 분석 - 2021년 이후 최고, 최저 종가 출력하기
min_price = data['Close'].min()
max_price = data['Close'].max()

print('최고 종가:',max_price)
print('최고 종가:',min_price)
```

```python
# D. 데이터 분석 - 2021년 이후 월 별 평균 종가 출력하기
data.groupby(data['Date'].dt.strftime('%Y-%m'))['Close'].mean().plot()

plt.title('Monthly Average Cllose Price')
plt.xlabel('Date')
plt.ylabel('Average Close Price')
plt.xticks(rotation = 45)
plt.show()
```

```python
data_2022 = data[data['Date'].dt.year == 2022]
plt.plot('Date', 'High', data=data_2022)
plt.plot('Date', 'Low',  data=data_2022)
plt.plot('Date', 'Close', data=data_2022)

plt.title('High, Low, and Close Prices since Janurary 2022')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation = 45)
plt.show()
```
